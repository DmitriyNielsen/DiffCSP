```python
import os
import pandas as pd
import random

# Define the directory
source_dir = "/Users/dmitriynielsen/GitHub/DiffCSP/CHILI-3K_CIFs"

# Get a list of all the .cif files
files = [f for f in os.listdir(source_dir) if f.endswith('.cif')]

# Read each .cif file and store its content as a row in a DataFrame
data = []
for file in files:
    with open(os.path.join(source_dir, file), 'r') as f:
        content = f.read()
        data.append([file, content])

df = pd.DataFrame(data, columns=['filename', 'content'])

# Shuffle the DataFrame
df = df.sample(frac=1).reset_index(drop=True)

# Split the DataFrame into train, test, and eval
train_df = df[:int(len(df)*0.7)]
test_df = df[int(len(df)*0.7):int(len(df)*0.85)]
eval_df = df[int(len(df)*0.85):]

# Save each set to a separate CSV file
train_df.to_csv("/Users/dmitriynielsen/GitHub/DiffCSP/CHILI-3K_CIFs/train_files.csv", index=False)
test_df.to_csv("/Users/dmitriynielsen/GitHub/DiffCSP/CHILI-3K_CIFs/test_files.csv", index=False)
eval_df.to_csv("/Users/dmitriynielsen/GitHub/DiffCSP/CHILI-3K_CIFs/eval_files.csv", index=False)
```
